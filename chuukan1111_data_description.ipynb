```python
using.describe()
```

```python
using = using[using["nendo"].isnull() == False]
```

```python
using["nendo"] = using["nendo"].astype("int64")
using = using.rename({"nendo":"fiscal_year"},axis=1)
```

```python
#年度別選挙数
using.groupby("fiscal_year").size().plot(kind="bar")
plt.xticks(rotation=30)
plt.grid(color='black', linestyle=':', linewidth=0.3)
plt.savefig("N_ele_by_year.png",dpi=500)
```

```python
#自治体別選挙数の
using.groupby("pres_pm_codes").size().value_counts().sort_index().plot(kind="bar")
plt.xticks(rotation=0)
plt.grid(color='black', linestyle=':', linewidth=0.3)
plt.xlabel("number of elections")
plt.savefig("N_ele_by_muni.png",dpi=500)
```
